```python
#import the needed libs
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
#set the default options to the web browser
options = webdriver.FirefoxOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
#create an instance of Chrome
browser = webdriver.Firefox(options=options)
```

```python
#get the landing page of the website
browser.get("https://paket.mk/")
#wait 5 seconds
browser.implicitly_wait(5)
#find all the <a></a> tags in the landing page
categories = browser.find_elements(By.TAG_NAME,"a")
#filter the ones that have "category" in their href
categories = [category.get_attribute("href") for category in categories if category.get_attribute("href").find("category")>0]
items = []
for category in categories:
    #get the page for each category
    browser.get(category)
    browser.implicitly_wait(5)
    #find all the <a></a> tags in the given category page
    products = browser.find_elements(By.TAG_NAME,"a")
    #filter the tags that have "product" in their href
    products = [product.get_attribute("href") for product in products if product.get_attribute("href").find("product")>0]
    print(products)
    for product in products:
        browser.get(product)
        browser.implicitly_wait(5)
        main = browser.find_element(By.ID,"portal-main")
        name = main.find_element(By.TAG_NAME,"a").text
        img = main.find_element(By.TAG_NAME,"img").get_attribute("src")
        spans = main.find_elements(By.TAG_NAME,"span")
        price = spans[0].text
        code = spans[3].text
        dist = spans[5].text
        items.append((code,name,price,dist,img))
    break#for whole page scraping delete this
```

```python
df = pd.DataFrame(items,columns=["code","name","price","distributor","image_src"])
df.to_csv("csv/items.csv")
df
```
